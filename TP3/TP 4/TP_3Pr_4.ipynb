##Implementación en Python


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqs

# Definir los coeficientes del numerador y del denominador de la función de transferencia
b = [0, 0, 1]  # Coeficientes del numerador (b0, b1, b2)
a = [1, 0.5, 1]  # Coeficientes del denominador (a0, a1, a2)

# Calcular la respuesta en frecuencia
w, h = freqs(b, a)

# Obtener magnitud y fase
magnitude = np.abs(h)
phase = np.angle(h)

# Crear figura y subgráficos
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Graficar la magnitud de la respuesta
axs[0].semilogx(w, 20 * np.log10(magnitude))  # Convertir magnitud a dB
axs[0].set_title('Respuesta en Magnitud')
axs[0].set_xlabel('Frecuencia [radianes / segundo]')
axs[0].set_ylabel('Magnitud [dB]')
axs[0].grid(True)

# Graficar la fase de la respuesta
axs[1].semilogx(w, np.degrees(phase))  # Convertir radianes a grados
axs[1].set_title('Respuesta en Fase')
axs[1].set_xlabel('Frecuencia [radianes / segundo]')
axs[1].set_ylabel('Fase [grados]')
axs[1].grid(True)

# Mostrar las gráficas
plt.tight_layout()
plt.show()
```

Explicación del Código:  
* Importamos las librerías necesarias.  
* Definimos los coeficientes de la función de transferencia. b son los coeficientes del numerador y a los del denominador.
* Usamos freqs de SciPy para calcular la respuesta en frecuencia del sistema para una gama de frecuencias angulares w.
* Calculamos la magnitud y la fase de la respuesta h.
* Finalmente, graficamos la magnitud y la fase utilizando Matplotlib. La magnitud es convertida a decibelios y la fase a grados para una mejor interpretación visual.
